```python
import matplotlib.pyplot as plt

# Number of threads
threads = [1, 2, 3, 4, 8, 16]

# Time taken in seconds
time_taken = [2585, 2385, 3101, 1349, 926, 645]

# Create a line graph
plt.plot(threads, time_taken, marker='o', linestyle='-', color='b')

# Set x-axis ticks to show only the specified values
plt.xticks(threads)

# Set axis labels and title
plt.xlabel('Number of Threads')
plt.ylabel('Time Taken (s)')
plt.title('Time Taken vs. Number of Threads')

# Show the plot
plt.show()
```
